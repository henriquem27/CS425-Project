```python
import pandas as pd 


teams_df= pd.read_csv('american_soccer_analysis_mls_xgoals_teams_2023-11-10.csv')
teams_df= teams_df.filter(['Team','Season','Games','ShtF','ShtA','GF','GA','Pts'])
teams_df['TeamID']=teams_df['Team']+teams_df['Season'].astype(str)
teams_df.to_csv('teams2.csv')

```

```python
player_df = pd.read_csv('american_soccer_analysis_mls_xgoals_players_2023-11-10.csv')

exp_player=player_df.filter(['Season','Player','Team','Position','Minutes','Shots','SoT','G'])

exp_player.replace('\s+','-',regex=True,inplace=True)
exp_player['Player_ID']=exp_player['Player']+'-'+exp_player['Season'].astype(str)
exp_player['Count']=exp_player.groupby('Player_ID').cumcount()+1
exp_player['Player_ID'] += '-'+exp_player['Count'].astype(str)

result_df = exp_player.merge(teams_df,on=['Team','Season'],how='left')

exp_player=result_df.filter(['Player_ID','Season','Player','TeamID','Position','Minutes','Shots','SoT','G'])

exp_player
#exp_player.to_csv('player2.csv')
```

```python
player_df = pd.read_csv('american_soccer_analysis_mls_xgoals_goalkeepers_2023-11-10.csv')

exp_gk = player_df.filter(['Player','Season','Team','Minutes','Shots Faced','Goals Conceded','Saves'])


exp_gk.replace('\s+','-',regex=True,inplace=True)
exp_gk['Player_ID']=exp_gk['Player']+'-'+exp_gk['Season'].astype(str)
exp_gk['Count']=exp_gk.groupby('Player_ID').cumcount()+1
exp_gk['Player_ID'] += '-'+exp_gk['Count'].astype(str)


result_df = exp_gk.merge(teams_df,on=['Team','Season'],how='left')

result_df
exp_gk=result_df.filter(['Player_ID','Season','Player','TeamID','Minutes','Shots Faced','Goals Conceded','Saves'])

exp_gk.to_csv('gk2.csv')

```

```python
games_df= pd.read_csv('american_soccer_analysis_mls_xgoals_games_2023-11-10.csv')

games_df=games_df.filter(['Date','Home','HG','Away','AG'])

games_df[['Year', 'Month', 'Day']] = games_df['Date'].str.split('-', expand=True)
games_df['Home_team_ID']=games_df['Home']+games_df['Year']

games_df['Away_team_ID']=games_df['Away']+games_df['Year']

games_df=games_df.filter(['Date','Home_team_ID','HG','Away_team_ID','AG','Year'])


games_df.to_csv('games2.csv')
```

```python
passes_df = pd.read_csv('american_soccer_analysis_mls_xpass_players_2023-11-10.csv')

passes_df['TeamID'] = passes_df['Team']+passes_df['Season'].astype(str)


passes_df.replace('\s+','-',regex=True,inplace=True)
result_df = passes_df.merge(exp_player,on=['TeamID','Season','Player'],how='left')

export_df=result_df.filter(['Player_ID','Passes','Minutes','Distance','Vertical','Score','Season'])
export_df.to_csv('passes2.csv')
```

```python

salaries_df=pd.read_csv('american_soccer_analysis_mls_salaries_players_2023-11-10.csv')


salaries_df['TeamID']=salaries_df['Team']+salaries_df['Season'].astype(str)

salaries_df.replace('\s+','-',regex=True,inplace=True)

result_df = salaries_df.merge(exp_player,on=['Season','Player','TeamID'],how='left')
export_df=result_df.filter(['Player_ID','Season','Base Salary','Guaranteed Compensation'])
export_df = export_df.drop_duplicates(subset='Player_ID', keep='first')
export_df
export_df.to_csv('salaries2.csv')
```
